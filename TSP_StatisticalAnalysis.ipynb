```
def merge_difficulty_set(df, prefix):
    
    cols = [col for col in df.columns if col.startswith(prefix)]
    
    def get_difficulty(row):
        for i, col in enumerate(cols, 1):
            val = row[col]
            if str(val).lower() == "yes":
                return i
        return None  

    df[f"{prefix}"] = df.apply(get_difficulty, axis=1)
    
    return df
    
df = merge_difficulty_set(df, "set1_5_difficulty")
df = merge_difficulty_set(df, "set6_10_difficulty")
df = merge_difficulty_set(df, "set11_15_difficulty")
df = merge_difficulty_set(df, "set16_20_difficulty")
df = merge_difficulty_set(df, "set21_25_difficulty")
df = merge_difficulty_set(df, "set26_30_difficulty")
df = merge_difficulty_set(df, "set31_35_difficulty")

print(df.to_string()) 
```

# STATISTICAL ANALISIS: 

License Status vs. Perceived Difficulty

- Null Hypothesis (Ho): There is no difference in the average perceived difficulty between participants with and without a driver's license. 

- Alternative Hypothesis (Ha): There is a difference in the average perceived difficulty between participants with and without a driver's license. 

```
from scipy.stats import ttest_ind, mannwhitneyu
import numpy as np

merged_df = pd.read_csv("merged_df.csv")

# Step 0: Clean column names to avoid hidden spaces
merged_df.columns = merged_df.columns.str.strip()

# Step 1: Define difficulty columns
difficulty_cols = [
    'set1_5_difficulty', 'set6_10_difficulty', 'set11_15_difficulty',
    'set16_20_difficulty', 'set21_25_difficulty', 'set26_30_difficulty', 'set31_35_difficulty'
]

# Step 2: Compute average difficulty
merged_df['avg_difficulty'] = merged_df[difficulty_cols].mean(axis=1)

# Step 3: Clean license status (map Yes/No to boolean)
if 'license' in merged_df.columns:
    merged_df['license'] = merged_df['license'].map({'Yes': True, 'No': False})
else:
    raise KeyError("Column 'license' not found in merged_df.")

# Step 4: Split groups
licensed = merged_df[merged_df['license'] == True]['avg_difficulty'].dropna()
unlicensed = merged_df[merged_df['license'] == False]['avg_difficulty'].dropna()

# Step 5: Test for normality through skewness
licensed_normal = np.abs(licensed.skew()) < 1
unlicensed_normal = np.abs(unlicensed.skew()) < 1

# Step 6: Run the statistical test
if licensed_normal and unlicensed_normal: 
    stat, p = ttest_ind(licensed, unlicensed, equal_var=False)
    test_used = "Independent t-test"
else: 
    stat, p = mannwhitneyu(licensed, unlicensed, alternative='two-sided')
    test_used = "Mann-Whitney U test"

# Step 7: Report results
print(f"Test used: {test_used}")
print(f"Licensed (n={len(licensed)}), mean difficulty: {licensed.mean():.2f}")
print(f"Unlicensed (n={len(unlicensed)}), mean difficulty: {unlicensed.mean():.2f}")
print(f"Statistic: {stat:.4f}")
print(f"P-value: {p:.4f}")

# Step 8: Interpret the results
if p < 0.05: 
    print("Result is statistically significant (p < .05).")
else: 
    print("Result is not statistically significant (p ≥ .05).")
```

Based on the results of this brief analysis we conducted, we noticed the p-value is just above the conventional threshold of 0.05. 

This means we cannot fully reject the null hypothesis with high confidence (the observed difference might be due to change). 

But it's very close to significance, and we can observe a trend where unlicensed participants found the tasks harder. 

```
sns.boxplot(data=merged_df, x='license', y='avg_difficulty')
plt.title("Perceived Difficulty by License Status")
plt.xlabel("Has Driver's License")
plt.ylabel("Average Difficulty (1-10)")
plt.xticks([0, 1], ['No', 'Yes'])
plt.show()
```

Now we aim to quantify how strong this difference is, by computing Cohen's d

```
def cohen_d(group1, group2): 
    #drop Nan values
    group1 = np.array(group1.dropna())
    group2 = np.array(group2.dropna())

    #sample sizes
    n1, n2 = len(group1), len(group2)

    #Pooled Standard Deviation (combining the standard deviations of two independent groups)
    pooled_std = np.sqrt(
        (((n1 - 1) * np.var(group1, ddof=1) + (n2 - 1) * np.var(group2, ddof=1)) / (n1 + n2 - 2))
    )

    #Cohen's d
    d = (np.mean(group1) - np.mean(group2)) / pooled_std
    return d


#we split both groups 
licensed = merged_df[merged_df['license'] == True]['avg_difficulty']
unlicensed = merged_df[merged_df['license'] == False]['avg_difficulty']

d_value = cohen_d(unlicensed, licensed)
print(f"Cohen's d: {d_value:.4f}")
```

Here we can appreciate how according to conventional thresholds (0.2 small, 0.5 medium, 0.8 big) , our effect size falls between the categories medium and large. 

This would suggest that even though the p-value (0.0515) was just above significance, the effect size is notable.  Suggesting a meaningful difference in perceived difficulty between participants with and without a license.

```
df_choices = pd.read_csv("processed_participants.csv")
df_licenses = pd.read_csv("merged_df2.csv")

# Normalize participant IDs for merging
df_choices['Participant_clean'] = df_choices['Participant'].str.lower()
df_licenses['Participant_clean'] = df_licenses['Participant_ID'].str.lower()

# Merge on the cleaned ID
merged_df = pd.merge(df_choices, df_licenses[['Participant_clean', 'license']], on='Participant_clean', how='inner')

# Drop NaN responses (only analyze actual choices)
merged_df = merged_df.dropna(subset=['Choice'])

# Create two separate contingency tables: one for license = True, one for license = False
contingency_license = pd.crosstab(
    merged_df[merged_df['license'] == True]['time_condition'],
    merged_df[merged_df['license'] == True]['Choice']
)

contingency_no_license = pd.crosstab(
    merged_df[merged_df['license'] == False]['time_condition'],
    merged_df[merged_df['license'] == False]['Choice']
)

from scipy.stats import chi2_contingency

# Run Chi-Square tests
chi2_license, p_license, _, _ = chi2_contingency(contingency_license)
chi2_no_license, p_no_license, _, _ = chi2_contingency(contingency_no_license)

contingency_license, chi2_license, p_license, contingency_no_license, chi2_no_license, p_no_license
```
